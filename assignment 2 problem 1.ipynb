```python
class BikeShop:
    bike_list = []
    
    @classmethod
    def add_bike(cls,obj): 
        cls.bike_list.append(obj)
        
    @classmethod
    def get_bike_list(cls):
        return BikeShop.bike_list
    
    @classmethod
    def get_bike_availability(cls):
        print("{}\t{}\t{}\t{}".format("Bike_id","Bike name","number of bikes","booked bikes"))
        count = 0
        bike_data = BikeShop.get_bike_list()
        for bike in bike_data:
            count+=1
            print("{}\t{:10}\t{}\t{}".format(count,bike.bike_name,bike.count,bike.booked))
        print("would you like to rent a Bike? press y for Yes and n for NO")
        ch = input()
        if ch == "Y" or ch == "y":
            return True
        else:
            return False
        
    
    def __init__(self,bike_name,count): 
        self.bike_name = bike_name
        self.count = count
        self.booked = 0
        BikeShop.add_bike(self)
    
    def update_bike_count(self,new_count): 
        self.count = new_count
        
        
```
